```python

plt.figure(figsize=(10,6))

plt.scatter(
    df['longitude'],
    df['latitude'],
    c=df['richter'],     # color = magnitude
    cmap='viridis',
    s=df['richter'] * 10 # size = magnitude
)

plt.colorbar(label='Richter Magnitude')

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earthquake Scatter Plot (Magnitude by Color & Size)')

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['longitude'], df['latitude'], c=df['richter'], cmap='hot', alpha=0.5, s=df['richter']**3)
plt.title('Earthquake Locations by Ritcher Scale')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Richter Magnitude')
plt.grid(linestyle=':')
plt.show()
```
