```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Expense_Categories = ['Groceries', 'Utilities', 'Transportation', 'Dining Out', 'Entertainment']
Amount_Spent = [500, 300, 200, 400, 250]

colors = ['blue', 'yellow', 'green', 'red', 'purple']

plt.figure(figsize=(10,7))
plt.barh(Expense_Categories, Amount_Spent, color=colors, height=0.7)

plt.xlabel('Amount Spent', fontsize=15)
plt.ylabel('Expense Categories',fontsize=15)
plt.title('Student Expenses by Category', fontsize=20, fontweight='bold')


plt.show()
```
