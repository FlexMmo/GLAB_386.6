### BarCharts

- `plt.bar()`
- `plt.barh()` - horizontal barchat
- Barcharts allow us to view the distribution of categorical values

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']
Student = [90, 60, 130, 190]


plt.bar(CoursesName, Student, color='Purple', width=0.5, zorder=3)

plt.xlabel('Course Name')
plt.ylabel('Student Reg Number')
plt.title('Number of Students Per Course')

plt.grid(True, axis='y', linestyle='--', alpha=0.5, zorder=0)

plt.show()
```

```python
plt.barh(CoursesName, Student, color='green') # uses y axis as the first and main axis
plt.xlabel('Student Reg')

plt.ylabel('Course Name')
plt.title('Number of Students Per Course')
plt.show()
```

```python
male = [90, 60, 130, 600]
female = [100, 30, 160, 690]

plt.bar(CoursesName, male, width=0.5, label='Male')
plt.bar(CoursesName, female, width=0.5, bottom=male, label='Female')

plt.title('Number of Students Per Course')
plt.xlabel('Courses Name')
plt.ylabel('Number of Students')
plt.legend()

plt.show()
```

```python
years = ['2016', '2017', '2018', '2019']
rice_price = [57, 50, 47, 30]
wheat_price = [43, 50, 53, 70]


plt.figure(figsize=(8,6))
plt.bar(years,rice_price, label='Rice')
plt.bar(years,wheat_price, label='Wheat',bottom=rice_price)
plt.title('Rice/Wheat Preferences Per Year', fontsize=20)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Preferences (%)', fontsize=15)
plt.legend(loc=2) # use loc parameter to change the position of your legend,eg 1,2,3,4 also upper right, upper left, center etc
plt.show()
```

```python
mustard =  np.array([10, 16, 13, 15])


plt.figure(figsize=(10,8))

plt.bar(years,rice_price,label='Rice',width=0.5)
plt.bar(years,wheat_price, label='Wheat', bottom=rice_price,width=0.5)
plt.bar(years,mustard, label='Mustard',bottom=np.array(rice_price) + np.array(wheat_price),width=0.5)


plt.title('Rice/Wheat/Mustard by Year', fontsize=20)
plt.xlabel('Years',fontsize=15)
plt.ylabel('Grains Frequency',fontsize=15)
plt.legend()


plt.show()
```

```python
# Creating a clustered Columns instead of stack
CoursesName = ['C Course', 'C++ Course', 'Java Course', 'Python Course']#
Male = [90, 60, 130, 600]
Female = [100, 30, 160, 690]

bar_width=0.35
x=np.arange(len(CoursesName))

print(x)


plt.bar(x - bar_width/2, Male, label='Male', width=bar_width)
plt.bar(x + bar_width/2, Female, label='Female',width=bar_width)
plt.xticks(x, CoursesName, rotation=45)
plt.legend()

plt.show()
```

```python
data = {'CoursesName': ['C Programming Course', 'C++ Programming Course', 'Java Programming Course', 'Python Programming Course'],
        'Student Group 1': [90, 60, 130, 600],
        'Student Group 2': [100, 30, 160, 690]}

df = pd.DataFrame(data)

bar_width=0.35
x=np.arange(len(CoursesName))

print(x)

plt.bar(x - bar_width/2, df['Student Group 1'], label='Student Group 1', width=bar_width)
plt.bar(x + bar_width/2, df['Student Group 2'], label='Student Group 2',width=bar_width)
plt.xticks(x, df['CoursesName'], rotation=45)
plt.title('Number of Students Per Course')
plt.xlabel('Course Name')
plt.ylabel('Number of Students')
plt.legend()



plt.show()
```
